```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
data
```

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=50)
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step') 
plt.show()
```

```python
plt.boxplot(data)  #Muestra rango intercuartil y la mediana
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)#Cambiar dirección, rellenar el rango inter cuartil patch_artist y enfocar la mediana notch
plt.show()
```

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True) #Quitar Outliers Showfliers 
plt.show()
```

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False) #Quitar Outliers Showfliers 
plt.show()
```

# Crear un gráfico tipo Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)#correlacion entre x y 
plt.show()
```

```python
plt.scatter(x,y, s=area)#Scatter con Área
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'o') # cambiar el marker
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'o', alpha= 0.5) # transparencia
plt.show()
```
